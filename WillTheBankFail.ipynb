```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def perform_bernoulli_trials(n_trials, probability):
  n_success = 0
  # perform trials again
  for i in range(n_trials):
    # choose random number between 0 and 1
    random_number = np.random.random()

    if random_number < probability:
      n_success += 1

  return n_success

def ecdf(data):
  """Compute ECDF for a one-dimensional array of measurements."""
  # Number of data points: n_data
  n_data = len(data)

  # x-data sorted for the ECDF: x
  x = np.sort(data)

  # y-data for the ECDF: y
  y = np.arange(1, (n_data + 1)) / n_data
  

  return x, y
```

```python
# seed random number generator
np.random.seed(42)
n = 1000
default_probability = 0.05
trials = 100
# initialize defaults array with n entries
n_defaults = np.empty(n)

# compute defaults probability for "trials" number of customers 
# n times with 5% default rate.
for i in range(n):
  n_defaults[i] = perform_bernoulli_trials(trials, default_probability)

```

```python
sns.set()

# _ = plt.hist(n_defaults)
# normed=True for normalizing frequency between 0 and 1
_ = plt.hist(n_defaults,  density=True)
_ = plt.xticks(ticks=np.arange(10))
_ = plt.xlabel('(# of defaults)/100')
_ = plt.ylabel('probability')

plt.show()
```

```python
x, y = ecdf(n_defaults)
# Plot the ECDF with labeled axes
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel("(# of defaults) / 100")
_ = plt.ylabel("cumulative frequency")

plt.show()
```
